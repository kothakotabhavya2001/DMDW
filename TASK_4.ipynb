3) Lets assume pre = np.array([19, 6, 11, 9]) and post = np.array([15, 11, 9, 8]) are the two 
Numpy array, plot this two array as side by side bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt 
X = ['Group A','Group B','Group C','Group D']
pre = np.array([19, 6, 11, 9])
post = np.array([15, 11, 9, 8]) 
X_axis = np.arange(len(X))
plt.bar(X_axis - 0.2, pre, 0.4, label = 'pre',color="blue")
plt.bar(X_axis + 0.2, post, 0.4, label = 'post',color="pink")
plt.xticks(X_axis, X)
plt.xlabel("PRE")
plt.ylabel("POST")
plt.legend()
plt.show()
```

4) Generate three NumPy arrays (X1, X2, X3) by getting 20 equidistance points between 0-100. Next convert this array into Y1, Y2, Y3 by applying sin, cos, and tan function over X1, X2, X3,respectively. Finally, plot these multiple lines over the same axis on a line plot graph.

```python
import math
import matplotlib.pyplot as plt
x1=np.arange(11,31)
x2=np.arange(31,51)
x3=np.arange(51,71)
y1=np.sin(x1)
y2=np.cos(x2)
y3=np.tan(x3)
plt.plot(x1,y1,label='X1')
plt.plot(x2,y2,label='X2')
plt.plot(x3,y3,label='X3')
plt.show()
```
